```python
# Agrupar client_id por Operations_ATM
df_op_atm=data.groupby('client_id')['Operations_ATM'].agg([sum,max,min, np.mean, np.count_nonzero, 'nunique'])
df_op_atm
```

```python
# Cambiar el nombre de las columnas
df_op_atm.columns = ['Op_ATM_' + str(col) for col in df_op_atm.columns]
df_op_atm
```

```python
# Poner client_id como columna
df_op_atm = df_op_atm.reset_index()
df_op_atm
```

```python
# Clone de DF
df_op_atm_2 = df_op_atm.copy()
df_op_atm_2.drop(['client_id'], inplace=True, axis=1)
df_op_atm_2
```

```python
# Genero otro DF para el cruce
df_op_homebanking = data.groupby('client_id')['Operations_HomeBanking'].agg([sum,max,min, np.mean, np.count_nonzero, 'nunique'])
df_op_homebanking
```

```python
# Cambiar el nombre de las columnas
df_op_homebanking.columns = ['Op_HomeBanking_' + str(col) for col in df_op_homebanking.columns]
df_op_homebanking
```

```python
# Definir el universo
universo = pd.DataFrame(data['client_id'].unique(), columns=['client_id'])
universo
```

```python
# Otra forma de generar un DF sin duplicados
universo_sin_duplicados = data[['client_id']].drop_duplicates()
universo_sin_duplicados
```

```python
# Genero el cruce
cruce = universo.merge(df_op_homebanking, how='left', on='client_id').merge(df_op_atm, how='left', on='client_id')
cruce.head(5).T
```

```python
# Operaciones en columnas. Ejemplo: Redondeo de un dato
cruce['Op_HomeBanking_mean'] = cruce['Op_HomeBanking_mean'].round()
cruce.head(5).T
```

```python
# Descripcion de un campo
cruce['Op_HomeBanking_mean_max'] = cruce['Op_HomeBanking_mean'] + cruce['Op_HomeBanking_max']
cruce.Op_HomeBanking_mean_max.describe()
```

```python
# Rellenar valores nulos con otro dato
data['Region'] = data['Region'].fillna('SIN INFORMACION')
data['Region']
```

```python
# Genero dato para union
region1 = data[data.Region == 'BUENOS AIRES'][['client_id', 'Region']]
region1['origen'] = '1'
region1
```

```python
# Genero otro dato para union
region2 = data[data.Region == 'CABA Centro/Norte'][['client_id', 'Region']]
region2['origen'] = '2'
region2
```

```python
# Union de datos
union = pd.concat([region1, region2])
union
```

```python
# Para saber de donde viene
union.origen.value_counts()
```
